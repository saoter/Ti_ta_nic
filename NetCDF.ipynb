Okay, NetCDF files are some kind of databases (not really), but kindof... in python you can use netCDF4 package, which connects to dataset and when you are done you need to close the connection. Here is example code to check variables and their format in your data.  

```python
pip install netCDF4
```

```python
import netCDF4

# use file path and Dataset function to read it in
f_path = 'wrfout_d02_2026-03-30_13:00:00'
dataset = netCDF4.Dataset(f_path, 'r') # r argument is for read only opion

# variables in your file, with shape and type (the same as standard pyton)
print("Variabels:")
for var_name in dataset.variables:
    print(f"- {var_name}: {dataset.variables[var_name].shape}, {dataset.variables[var_name].dtype}")

# meta data of the file or attributes of the data (normally time, location...etc...)
print("\nGlobal Attributes:")
for attr_name in dataset.__dict__:
    print(f"- {attr_name}: {dataset.__dict__[attr_name]}")

# close it at the end...
dataset.close()
```
